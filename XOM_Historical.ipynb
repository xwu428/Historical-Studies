Exxonmobile stock price

```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

# Download historical stock price data for ExxonMobil
xom = yf.download('XOM', start='1970-01-01', end='2022-02-20')

# Create a figure and axes for the graph
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the stock price over time
ax.plot(xom['Close'], color='blue', linewidth=2)

# Add markers for the global events mentioned above
ax.axvline(datetime.strptime('1973-10-01', '%Y-%m-%d'), color='red', linestyle='--', label='1973 OPEC oil embargo')
ax.axvline(datetime.strptime('1986-01-01', '%Y-%m-%d'), color='orange', linestyle='--', label='1986 oil price collapse')
ax.axvline(datetime.strptime('1991-08-02', '%Y-%m-%d'), color='green', linestyle='--', label='1991 Gulf War')
ax.axvline(datetime.strptime('2011-03-11', '%Y-%m-%d'), color='purple', linestyle='--', label='2011 Fukushima disaster')
ax.axvline(datetime.strptime('2020-03-01', '%Y-%m-%d'), color='brown', linestyle='--', label='2020 COVID-19 pandemic')

# Set the x-axis label
ax.set_xlabel('Year')

# Set the y-axis label
ax.set_ylabel('Stock Price (USD)')

# Set the title of the graph
ax.set_title('ExxonMobil Stock Price (1970-2022)')

# Add a legend to the graph
ax.legend(loc='best')

# Show the graph
plt.show()
```
